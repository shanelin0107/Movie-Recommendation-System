# **Extract Info from Genres**

```python
# Create function to extract values ('name') from a dict
def get_features(lst):
    feat = []
    for i in ast.literal_eval(lst):
        feat.append(i['name'])
    return feat
```

```python
movies1['genres_names'] = movies1['genres'].apply(get_features)
movies1[['genres','genres_names']].head()
```

# **Preprocess Data (overview)**

```python
# Create function to remove punctuation
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

# Remove punctuation from data
movies1['overview_text'] = movies1['overview'].apply(lambda x: remove_punct(x))
movies1.head()
```

```python
#tokenize the data
def tokenize(text):
    # W+ means that either a word character (A-Za-z0-9_) or a dash (-) can go there.
    tokens = re.split('\W+', text)
    return tokens

# Tokenize and lowercase data 
movies1['overview_text_tokenized'] = movies1['overview_text'].apply(lambda x: tokenize(x.lower())) 
movies1.head()
```

```python
stopwords = nltk.corpus.stopwords.words('english')     # All English Stopwords
stopwords
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopwords]
    return text

# Remove stop words from data
movies1['overview_text_nostop'] = movies1['overview_text_tokenized'].apply(lambda x: remove_stopwords(x))
movies1.head()
```

```python
wn = nltk.WordNetLemmatizer()

# Create function to apply lematizer
def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

# Apply lemmatizer
movies1['overview_text_lemmatized'] = movies1['overview_text_nostop'].apply(lambda x: lemmatizing(x))
movies1.head(10)
```

```python
movies1['combination'] = movies1['genres_names'] + movies1['overview_text_lemmatized']
movies1.head()
```

# **Vectorize Data - TF-IDF**

```python
tfidf = TfidfVectorizer(analyzer='word')
```

```python
#TFIDFVectorizer is unable to apply a list type dataset, so we convert the "combination" list into a string dataset
movies1['string'] = [' '.join(map(str, l)) for l in movies1['combination']]
```

```python
movies1.head()
```

```python
overview_matrix = tfidf.fit_transform(movies1['string'])
overview_matrix
```

```python
cosine_similarities = cosine_similarity(overview_matrix)
cosine_similarities.shape
```

```python
cosine_similarities
```

```python
# Print sample cosine_similarities row
cosine_similarities[15]
```

```python
# Print top 10 indicies from sample cosine_similarities row in descending sorted order
#    argsort performs an indirect sort on a numpy array
#    It returns an array of indicies in sorted order
cosine_similarities[15].argsort()[:-10:-1]
```

```python
%%time

# Store the names of the 50 most similar movies for each movie in our dataset
similarities = {}
for i in range(len(cosine_similarities)):
    # Sort each element in cosine_similarities and get the indexes of the 50 most similar movies
    similar_indices = cosine_similarities[i].argsort()[:-50:-1]
    
    # Store in similarities each name of the 50 most similar movies
    # Note: the first one is the same movie
    similarities[movies1['title'].iloc[i]] = [(cosine_similarities[i][x], movies1['title'][x],)
                                           for x in similar_indices][1:]
```

# **Test The Recommendation System**

```python
# Use the similarity scores (generated above) to access the most similar items and make recommendations
# Create Content based recommender class
class ContentBasedRecommender:
    def __init__(self, matrix):
        self.matrix_similar = matrix

    def _print_message(self, movie, recom_movie):
        rec_items = len(recom_movie)
        
        print(f'The {rec_items} recommended movies for {movie} are:')
        for i in range(rec_items):
            print(f"{i+1}: {recom_movie[i][1]} with {round(recom_movie[i][0], 3)} similarity score") 
        
    def recommend(self, recommendation):
        # Get movie to find recommendations for
        movie = recommendation['movie']
        # Get number of movies to recommend
        number_movies = recommendation['number_movies']
        # Get the number of movies most similars from matrix similarities
        recom_movie = self.matrix_similar[movie][:number_movies]
        # print each item
        self._print_message(movie=movie, recom_movie=recom_movie)
```

```python
# Instantiate Content based recommender class with the precomputed similarities
recommedations = ContentBasedRecommender(similarities)
```

```python
# Pick a movie and make a recommendation
recommendation = {
    "movie": movies1['title'].iloc[10],
    "number_movies": 4 
}
recommedations.recommend(recommendation)
```

```python
# Pick another movie and make a recommendation
recommendation2 = {
    "movie": movies1['title'].iloc[120],
    "number_movies": 5 
}
recommedations.recommend(recommendation2)
```
